#Project 1: Road Traffic Severity Classification Description: This data set is collected from Addis Ababa Sub-city police departments for master's research work. The data set has been prepared from manual records of road traffic accidents of the year 2017-20. All the sensitive information has been excluded during data encoding and finally it has 32 features and 12316 instances of the accident. Then it is preprocessed and for identification of major causes of the accident by analyzing it using different machine learning classification algorithms.

task 3 - Feature Engineering and Modeling

```python
import pandas as pd
import matplotlib.pyplot as plt
#from autoviz.AutoViz_Class import AutoViz_Class
#from autoviz import data_cleaning_suggestions
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn import metrics
#import missingno as msno
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('RTA Dataset_encoded.csv')
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df.head()
```

```python
#select All features, except the target variable
labels = df.columns.to_list()
labels.remove('Accident_severity_encoded')
labels
```

```python
df.columns
```

```python
#prepare the datasets to apply in the models
X = df[labels]
y = df['Accident_severity_encoded']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
#predict using XGBClassifier

model = XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
#Evaluate the model
accuracy = metrics.accuracy_score(y_test, y_pred)
f1_score = metrics.f1_score(y_test, y_pred, average='weighted')
print("Accuracy: %.2f%%" % (accuracy * 100.0))
print("f1 score: %.2f%%" % (f1_score * 100.0))
```

```python
#predict using randomforest
rfcl = RandomForestClassifier()

rfcl.fit(X_train,y_train)

y_pred_rf = rfcl.predict(X_test)
```

```python
accuracy2 = metrics.accuracy_score(y_test, y_pred_rf)
f1_score2 = metrics.f1_score(y_test, y_pred_rf, average='weighted')
print("Accuracy: %.2f%%" % (accuracy2 * 100.0))
print("f1 score: %.2f%%" % (f1_score2 * 100.0))
```

Task 4. Hyperparameter tuning

```python
#tuning with GridSearchCV
forest_params = [{'max_depth': list(range(10, 15)), 'max_features': list(range(0,14))}]

clf = GridSearchCV(rfcl, forest_params, cv = 10, scoring='accuracy')

clf.fit(X_train, y_train)

print(clf.best_params_)

print(clf.best_score_)
```
